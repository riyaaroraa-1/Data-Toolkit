# Theory

1. What is NumPy, and why is it widely used in Python?

     Ans. NumPy is a popular Python library used for numerical and mathematical operations. It provides support for arrays, matrices, and many useful mathematical functions.

     NumPy is widely used because it is fast, efficient, and easy to use. It performs calculations much faster than standard Python lists and helps in handling large amounts of data. It also supports multi-dimensional arrays and operations like sorting, reshaping, indexing, and statistical calculations.



2. How does broadcasting work in NumPy?

     Ans. Broadcasting in NumPy is a feature that allows arrays of different shapes or sizes to perform mathematical operations together without manually resizing them.

     NumPy automatically adjusts the smaller array to match the shape of the larger array during the operation. This makes calculations faster and reduces the need for writing complex loops.

3. What is a Pandas DataFrame?
    
     Ans. Pandas DataFrame is a two-dimensional data structure used to store and organize data in rows and columns, similar to a table or spreadsheet.

     A DataFrame can contain different types of data such as numbers, text, and dates. It provides various functions for data analysis, filtering, sorting, cleaning, and manipulation.

     

4. Explain the use of the groupby() method in Pandas.

     Ans. The `groupby()` method in Pandas is used to group data based on one or more columns and perform operations on each group separately. It helps in data analysis by summarizing, aggregating, or transforming data easily.

     For example, if a dataset contains employee departments and salaries, `groupby()` can be used to calculate the average salary of each department.

5. Why is Seaborn preferred for statistical visualizations?

     Ans. Seaborn is preferred for statistical visualizations because it provides attractive and informative graphs with less code. It is built on top of Matplotlib and offers better default styles and easier functions for creating statistical plots such as heatmaps, bar plots, box plots, and pair plots.

     Seaborn also works well with Pandas DataFrames, making data analysis and visualization simpler and more efficient. It helps users easily understand patterns, relationships, and distributions in data.

6. What are the differences between NumPy arrays and Python lists?

     Ans. NumPy arrays and Python lists differ in performance, functionality, and memory usage.

      *   NumPy arrays are faster and more efficient because they store data of the same type in a continuous memory location. They support mathematical and vectorized operations directly.

     *  Python lists can store different data types together, but they are slower for numerical computations and use more memory.

     NumPy arrays are mainly used for scientific computing and data analysis, while Python lists are used for general-purpose programming.



7. What is a heatmap, and when should it be used?

     Ans. A heatmap is a data visualization technique that represents values using different colors. The color intensity helps users quickly understand patterns, relationships, or variations in the data.

     Heatmaps are commonly used to show correlation matrices, missing data, performance comparisons, and large datasets where patterns are difficult to identify in tables.

     For example, in data analysis,  a heatmap can be used to display the correlation between variables, where darker or brighter colors indicate stronger relationships.

8. What does the term “vectorized operation” mean in NumPy?

     Ans. A vectorized operation in NumPy means performing operations on entire arrays at once without using explicit loops. NumPy internally uses optimized C code, making these operations much faster and more efficient than traditional Python loops.

9. How does Matplotlib differ from Plotly?

     Ans. Matplotlib and Plotly are both data visualization libraries in Python, but they differ in features and usage.

      * Matplotlib is mainly used for creating static graphs and provides detailed control over plot customization. It is widely used for basic charts like line graphs, bar charts, and histograms.

      * Plotly is used for creating interactive and dynamic visualizations. It allows users to zoom, hover, and interact with charts directly in web browsers.


10. What is the significance of hierarchical indexing in Pandas?

     The main significance of hierarchical indexing is that it enables users to work with higher dimensional data in a two-dimensional DataFrame. It also makes data selection, grouping, and filtering easier.

15. How does NumPy handle multidimensional arrays?

     Ans. NumPy handles multidimensional arrays using a powerful data structure called `ndarray` (N-dimensional array). These arrays can store data in multiple dimensions such as 1D, 2D, 3D, and higher dimensions while maintaining fast computation and efficient memory usage.

     For example:

     * A 1D array represents a list of elements.

     * A 2D array represents rows and columns like a matrix.

     * A 3D array can represent multiple matrices or complex datasets.

     NumPy provides functions to create, reshape, index, slice, and perform mathematical operations on multidimensional arrays easily. It also supports vectorized operations, which allow calculations on entire arrays without using loops, making computation faster and more efficient.

17. Explain the difference between apply() and map() in Pandas?

     Ans. `map()` and `apply()` are both used in Pandas for data transformation, but they have different purposes.

     * `map()` is mainly used with a Series (single column) and performs element-wise operations. It is useful for simple value replacements or transformations.
     * `apply()` is more flexible and can be used on both Series and DataFrames. It applies a function across or columns and is suitable for more complex operations.

     In simple terms, `map()` is best for single-column element-wise changes, while `apply()` is used for broader and more advanced data operations.

20. What is the role of a pivot table in Pandas?

      The `pivot_table()` function allows users to compare different categories of data and generate insights quickly. It is commonly used for data analysis, reporting, and creating summaries from complex datasets.


21. Why is NumPy’s array slicing faster than Python’s list slicing?

     Ans. NumPy is faster than Python's list slicing because NumPy arrays are stored in contiguous memory locations and are implemented in optimized C code. This allows operations to be performed efficiently without creating unnecessary copies of data.
     
      In contrast, Python lists store references to objects and require more memory and processing time. NumPy slicing often returns a view of the original array instead of creating a new copy, which improves speed and memory efficiency.

1. How do you create a 2D NumPy array and calculate the sum of each row?

```python
# Create a 2D NumPy array
data = np.array([[2,4,6],[1,3,5],[1,7,9]])

# Calculate the sum of each row
row_sum = np.sum(data, axis = 1)
print("Sum:",row_sum)
```

2. Write a Pandas script to find the mean of a specific column in a DataFrame.

```python
# Creating a sample data
data1 = {'Name':['Arya','Tanvi','Shane','Evan','Mike','Kartik','Manik'],
'Age':[20,18,15,25,24,28,27]}

df = pd.DataFrame(data1)

# calculating the mean of age column
print("Mean age:",round(df['Age'].mean()))
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
tips = sns.load_dataset('tips')

numeric_data = tips.select_dtypes(include = 'number')

# calculating correlation matrix
corr = numeric_data.corr()

# heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr , annot = True )
plt.title('Correlation Matrix - Tips')
plt.show()
```

8. Create a line plot with multiple lines using Matplotlib.


```python
years = [2021,2022,2023,2024,2025]
productA_sales = [50,60,63,72,80]
productB_sales = [45,55,59,60,75]

plt.plot(years,productA_sales,label ='Product A', color = 'purple',marker ='o')
plt.plot(years,productB_sales, label ='Product B',color = 'green', marker = 's')

plt.xlabel('Years')
plt.ylabel('Sales(in thousands)')
plt.xticks(years)
plt.title('Yearly Sales Comparison')
plt.grid(True)
plt.legend()
plt.show()
```

9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.


```python
sales_data = {'Product':['Laptop','Monitor','Smartphone','Laptop','Printer','Tablet'],
              'Sales':[95000,40000,35000,90000,45000,64000]}

sales_df = pd.DataFrame(sales_data)
print("Original DataFrame:\n",sales_df)

# Filtering rows where sales are greater than 50000
threshold = 50000
high_sales_df = sales_df[sales_df['Sales'] > threshold]

print("\nProducts with Sales > 50000:\n",high_sales_df)

```

11. Perform matrix multiplication using NumPy.


```python
A = np.array([[1,3,5],
              [2,4,6],
              [1,2,3]])

B = np.array([[6,8,10],
              [7,9,11],
              [3,2,1]])

C = np.dot(A,B)# Matrix multiplication
print("Result:\n",C)
```
